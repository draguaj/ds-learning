## Задание.

Создайте случайную величину из [любого выбранного вами распределения](https://docs.scipy.org/doc/scipy/reference/stats.html), для разных значений `n` сгенерируйте 1000 выборок размера `n`. Постройте гистрограммы средних этих выборок и сделайте выводы.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```
n1 = 5
rv = sts.genlogistic(n1) # Обобщенная логистическая непрерывная случайная величина
sample = rv.rvs(size=1000)
```

```
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```
n2 = 25
rv = sts.bradford(n2)  # Непрерывная случайная величина Брэдфорда
sample = rv.rvs(size=1000)
```

```
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```
# n1 = 5                # Размеры выборок
# n2 = 25

a1 = np.array([])     # Массивы для сохранения средних по выборкам
a2 = np.array([])

samples_count = 1000  # Количество повторов генерации

for number in range(samples_count):
    sample1 = rv.rvs(size=n1)               # На каждом цикле генерируем выборки указанных размеров
    sample2 = rv.rvs(size=n2)
    a1 = np.append(a1, sample1.mean())      # Считаем среднее по каждой выборке и добавляем в массив средних
    a2 = np.append(a2, sample2.mean())
```

```
label_1 = 'sample of ' + str(n1) + ' histogram'   # Подписи для графиков
label_2 = 'sample of ' + str(n2) + ' histogram'
plt.hist(a1, density=True, label=label_1)         # Строим графики
plt.hist(a2, density=True, label=label_2)

plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
```

##Выводы
- Среднее значение распределения средних значений выборки размера n1 приближается к исходному среднему значению исходного распределения. То же самое наблюдается для выборки размера n2.
- Распределение средних значений для выборки размера n2 (большей выборки) имеет **меньшую дисперсию**, что соответствует закону больших чисел. Это означает, что средние значения выборки размера n2 более стабильны и менее изменчивы.
- Эксперимент иллюстрирует, как увеличение размера выборки способствует устойчивости и точности оценки среднего значения.
