### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

##Решение
A1 и A2 попарно независимы, потому что исход каждого из них не влияет на исход другого события. Вероятность того, что монеты 1 и 2 упадут одной и той же стороной, не зависит от того, как упадут монеты 2 и 3.
Аналогично можно сказать о попарной независимости событий A1 и A3, а также A2 и A3.

Рассмотрим независимость в совокупности. Допустим мы знаем, что событие А1 уже произошло и вероятность события А2 становится зависимым от А1. Вероятность события А2 изменяется.
Таким образом, эти три события (A1, A2, A3) не являются независимыми в совокупности, так как знание одного из событий оказывает влияние на вероятность других событий.

### Задание 3


Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```
import numpy as np
from scipy.stats import norm, expon, t
import matplotlib.pyplot as plt
```

##Нормальное распределение

```
parameters = {
    'norm': [
        {'loc': 0, 'scale': 1},
        {'loc': 0, 'scale': 0.5},
        {'loc': 2, 'scale': 1},
        {'loc': -2, 'scale': 2},
        {'loc': 0, 'scale': 2}
    ]}

x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
plt.title('Нормальное распределение')
for params in parameters['norm']:
    plt.plot(x, norm.pdf(x, **params), label=str(params))
plt.legend()
```

###Вывод
Большее значение scale делает распределение шире.
Увеличение loc сдвигает пик распределения вправо, а уменьшение - влево.

##Экспоненциальное распределение

```
parameters = {
    'expon': [
        {'scale': 1},
        {'scale': 0.5},
        {'scale': 2},
        {'scale': 0.2},
        {'scale': 5}
    ]}

x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 2)
plt.title('Экспоненциальное распределение')
for params in parameters['expon']:
    plt.plot(x, expon.pdf(x, **params), label=str(params))
plt.legend()
```

###Вывод
Меньшее значение scale делает распределение более "острым".

##Распределение Стьюдента

```
parameters = {
    't': [
        {'df': 2},
        {'df': 5},
        {'df': 10},
        {'df': 20},
        {'df': 50}
    ]}

x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 3)
plt.title('Распределение Стьюдента')
for params in parameters['t']:
    plt.plot(x, t.pdf(x, **params), label=str(params))
plt.legend()
```

###Вывод
С ростом df(степени свободы) распределение становится более концентрированным вокруг среднего значения.
