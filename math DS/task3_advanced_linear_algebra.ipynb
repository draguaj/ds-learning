### Задание 1

Дан объект в $2D$ пространстве

````
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````
# Деление каждой координаты на 2
scaled_A = A / 2

x = scaled_A[:, 0]
y = scaled_A[:, 1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
# Угол поворота в градусах
angle_degrees = 130
````

````
# Преобразование угла в радианы
angle_radians = np.radians(angle_degrees)
````

````
# Матрица поворота
rotation_matrix = np.array([[np.cos(angle_radians), -np.sin(angle_radians), 0],
                            [np.sin(angle_radians), np.cos(angle_radians), 0],
                            [0, 0, 1]])
````

````
# Применение матрицы поворота к набору точек A
rotated_A = A.dot(rotation_matrix) / 2

x = rotated_A[:, 0]
y = rotated_A[:, 1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
# Матрица отражения относительно прямой y=x
_matrix = np.array([[0, 1, 0],
                    [1, 0, 0],
                    [0, 0, 1]])

# Применение матрицы отражения к набору точек A
reflected_A = A.dot(_matrix, A)

x = reflected_A[:, 0]
y = reflected_A[:, 1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
m = np.array([[1, 2],
              [2, 3]])
print(m)
````

````
ev, eW = np.linalg.eigh(m)
L=np.diag(ev)
np.dot(eW.T, eW) #матрицы ортогональны
np.dot(np.dot(eW, L), eW.T) #получим исходную матрицу m
````
