### Задание 1


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.


```
import sympy as sym
from sympy import diff, symbols, cos, sin
import numpy as np

x = sym.Symbol('x')
func = x**5 + 4 * sym.sin(2 * x) + sym.cos(3 * x + 3)

first = sym.diff(func, x)  # Вычисление первой производной
second = sym.diff(first, x)  # Вычисление второй производной
x_value = 1

first_x1 = first.evalf(subs={x: x_value}) # Значение первой производной
second_x1 = second.evalf(subs={x: x_value}) # Значение второй производной
print("Первая производная при x=1:", first_x1)
print("Вторая производная при x=1:",second_x1)
```

### Задание 2

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```
from sympy import *

x = Symbol('x')

y = sin(2*x+1)**5
y_first = diff(y, x)
y_second = diff(y_first, x)

from sympy.plotting import plot

plot(y_first, y_second, ('x', -5, 5, 100))
```
